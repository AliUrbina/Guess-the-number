###**Fix NULLS**

```python
#To fix all the NULLs in the dataset, we need to identify how many of this exist in the data.
df.isnull().sum()
```

###**Remove inconsistent data**

```python
#Remove departments with value = 0. first we need to ask my data frame, this information
df.query("Valor==0 | Condominio==0")
```

```python
#Identify all the index. This will help us to remove all the 0's at once
df.query("Valor==0 | Condominio==0").index
```

```python
#For this we need to creat a filter
df_remover = df.query("Valor==0 | Condominio==0").index
```

```python
#With .drop and inplace=True we are telling the dataframe to store everythin in a variable withou the need to review it again.
df.drop(df_remover, axis=0, inplace=True)
```

```python
df.query("Valor==0 | Condominio==0").index
```

```python
#Now, since we are working just with departments in this dataframe, maybe the column "Tipo", doesn't have sense anymore, so we can removeit
df.Tipo.unique()
```

```python
#Removing the column Tipo with .drop
df.drop("Tipo", axis=1, inplace=True)
df.sample(10)
```

```python
#We need to import again the complete Database
dev_data = pd.read_csv(url, sep=';')
dev_data.shape
```

```python
#Create Columns in pandas is with the column name = dataframe name["Columns name"]
dev_data["Valor_mensual"] = dev_data["Valor"] + dev_data["Condominio"]
dev_data.head()
```

###2. Annual value

```python
dev_data["Valor_Anual"] = dev_data["Valor_mensual"] * 12 + dev_data["Impuesto"]
dev_data.tail()
```

```python
from numpy import astype
#creat column "Description" + tyoe of property, concatenar + Column1 + blank space + column 2

'''
dev_data["Descripcion"] = dev_data["Tipo"] + " en la colonia  " +dev_data["Colonia"] \
                          + " con " + dev_data["Habitaciones"] + " cuarto(s) y " + \
                          dev_data["Garages"] + " plazas de estacionamiento. "

'''
#Since the last code give an error that we onlyu can concatenate str not int we need to fix it
#with astype() metod

dev_data["Descripcion"] = dev_data["Tipo"] + " en la colonia  " +dev_data["Colonia"] \
                          + " con " + dev_data["Habitaciones"].astype(str) + " cuarto(s) y " + \
                          dev_data["Garages"].astype(str) + " plazas de estacionamiento. "
```

###2. Create a column to know ig the property has suites or not

```python
#using our data frame we need to create a new column and in this column we are applying a fuctions
#to determinate if the property has suits or not, if this have value es 1 and 0 is no with Lambda

dev_data["Tiene_Suit"] = dev_data["Suites"].apply(lambda x: "Si" if x > 0 else "No")
```
